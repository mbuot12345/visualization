```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
USA_Housing=pd.read_csv(r"C:\Users\Mbuotidem urom\Downloads\archive (3)\USA_Housing.csv")
USA_Housing.head(10)
```

```python
x1 = [1,2,3,5,7,8,11]
y1 = [2,4,1,10,14,13.6,6]
plt.plot(x1,y1)
plt.xlabel('X-axis')# naming the x axis
plt.ylabel('y-axis')
plt.title('My First Line Plot')
plt.legend('line')
plt.show()
```

```python
# Generatimg x and y points to plot
x1 = np.arange(0,10*(np.pi),0.1)
y1 = np.cos(x1)# cosine function from numpy library
plt.plot(x1,y1,label ='cosine')
plt.legend()
plt.xlabel('Angle')
plt.ylabel('Amplitude')
x2 = np.arange(0,10*(np.pi),0.1)
y2 = np.cos(x2)*0.1*np.arange(10*(np.pi),0,-0.1)
plt.plot(x2,y2,label = 'Decaying cosine')
plt.legend()
plt.title('Two functions on the same graph')
plt.show()
```

```python
# plotting Avg. Area income vs price
plt.plot(USA_Housing['Avg. Area Income'],USA_Housing['Price'],color='red',marker='o')
plt.title('Avg. Area Income vs Price',fontsize=14)
plt.xlabel('Avg. Area Income',fontsize=14)
plt.ylabel('Price',fontsize=14)
plt.grid(True)
plt.show()
```

```python
# Adjusting the figure size using rcParams function 
plt.rcParams['figure.figsize']=[12,8]
plt.scatter(USA_Housing['Avg. Area Income'],USA_Housing['Price'],color='red',marker='o')
plt.title('Avg. Area Income vs Price',fontsize=14)
plt.xlabel('Avg. Area Income',fontsize=14)
plt.ylabel('Price',fontsize=14)
plt.grid(True)
plt.show
```
